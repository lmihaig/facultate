```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1
mean = 0
std = 1
size = 10000
dist_uni = np.random.normal(mean, std, size)

mean = [0, 0]
cov = [[1, 3 / 5], [3 / 5, 2]]
dist_multi = np.random.multivariate_normal(mean, cov, size)


fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].hist(dist_uni, bins=100, density=True, label="normal")
ax[1].scatter(dist_multi[:, 0], dist_multi[:, 1], label="normal")
plt.show()
```
