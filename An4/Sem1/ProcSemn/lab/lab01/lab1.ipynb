```python
import matplotlib.pyplot as plt
import numpy as np
import humanize

plt.rcParams["figure.figsize"] = [8, 5]
plt.rcParams["figure.dpi"] = 140
```

```python
# 1
def cos_wave(omega, t, amp):
    return np.cos(omega * np.pi * t + np.pi / amp)


x = lambda t: cos_wave(520, t, 3)
y = lambda t: cos_wave(280, t, -3)
z = lambda t: cos_wave(120, t, 3)
```

```python
# a + b
dreapta = 0.15
interval = np.linspace(0, dreapta, int(dreapta / 0.0005))
fig, axs = plt.subplots(3)
fig.suptitle("Semnale initiale")
axs[0].plot(interval, x(interval), color="red")
axs[1].plot(interval, y(interval), color="green")
axs[2].plot(interval, z(interval), color="blue")

axs[0].set_xlabel("Timp (s)")
axs[0].set_ylabel("x(t)")
axs[1].set_ylabel("y(t)")
axs[2].set_ylabel("z(t)")

for ax in axs:
    ax.set_xlim(0, dreapta)

plt.show()
```

```python
# c
sampled_interval = np.linspace(0, dreapta, 200)
fig, axs = plt.subplots(3)
fig.suptitle("Semnale esantionate")
axs[0].stem(sampled_interval, x(sampled_interval), linefmt="r-")
axs[1].stem(sampled_interval, y(sampled_interval), linefmt="g-")
axs[2].stem(sampled_interval, z(sampled_interval), linefmt="b-")

axs[0].set_xlabel("Time (s)")
axs[0].set_ylabel("x[t]")
axs[1].set_ylabel("y[t]")
axs[2].set_ylabel("z[t]")

for ax in axs:
    ax.set_xlim(0, dreapta)

plt.show()
```

```python
# 2
# a
freq_signal = 400
samples = 1600
t = np.linspace(0, 1 / freq_signal, samples)
plt.stem(t, np.sin(2 * np.pi * freq_signal * t))
plt.title("Semnal sinusoidal 400Hz")
plt.xlabel("Timp (s)")
plt.ylabel("a[t]")
plt.grid(True)
plt.show()
```

```python
# b
freq_signal = 800
duration = 3
t = np.linspace(0, duration, 3 * freq_signal)
plt.stem(t, np.sin(2 * np.pi * freq_signal * t))
plt.title("Semnal sinusoidal 800Hz")
plt.xlabel("Timp (s)")
plt.ylabel("b[t]")
plt.grid(True)
plt.show()
```

```python
# c
freq_signal = 240
duration = 4 / 240
t = np.linspace(0, duration, 100)
plt.stem(t, freq_signal * t - np.floor(freq_signal * t))
# formula generala de pe wikipedia da valori [-1,1] ?
# plt.stem(t, 2 * (t - np.floor(1 / 2 + t)))
plt.title("Semnal Sawtooth")
plt.xlabel("Timp (s)")
plt.ylabel("c[t]")
plt.grid(True)
plt.show()
```

```python
# d
freq_signal = 300
duration = 2 / 300
t = np.linspace(0, duration, 100)
plt.stem(t, np.sign(np.sin(2 * np.pi * freq_signal * t)))
plt.title("Semnal Square")
plt.xlabel("Timp (s)")
plt.ylabel("d[t]")
plt.grid(True)
plt.show()
```

```python
# e
I = np.random.rand(128, 128)
plt.imshow(I, cmap="gray")
plt.title("Semnal 2D aleator")
plt.show()
```

```python
# f
x = 128
I = np.zeros((x, x))

for i in range(x):
    for j in range(x):
        if (i + j) % 10 == 0:
            I[i, j] = 1

plt.imshow(I, cmap="gray")
plt.title("Semnal 2D diagonale")
plt.show()
```
